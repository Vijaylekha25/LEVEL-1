```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
ur1="https://drive.google.com/file/d/1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y/view?usp=drive_link"
ur1='https://drive.google.com/uc?id=' + ur1.split('/')[-2]
df=pd.read_csv(ur1)
```

##### Task 1: Top Cuisines

##### Determine the top three most common cuisines in the dataset.

```python
print(df.shape)
print(df.columns)
```

```python
print(df['Cuisines'])
```

```python

Cuisines_count = df['Cuisines'].str.split(',').explode().value_counts()
print(Cuisines_count)
top_cuisine = Cuisines_count.head(3)
print("The Top 3 Cuisines are:", top_cuisine)
```

```python
colours = ['green', 'orange', 'red'] 
plt.bar(top_cuisine.index, top_cuisine.values, color=colours)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top Three Cuisines')
plt.show()
```

##### Calculate the percentage of restaurants that serve each of the top cuisines

```python

total_restaurant = len(df)
print(total_restaurant)
top_cuisine10 = Cuisines_count.head(5)
percentage = (top_cuisine10 / total_restaurant) * 100
print("The Market share of Top 5 Cuisines are:", percentage)
```

```python
plt.bar(top_cuisine10.index, percentage.values, color=colours)
plt.xlabel('Cuisines_Name')
plt.ylabel('percentage')
plt.title('Top five Cuisines %')
plt.figure(figsize=(12,6))
plt.show()
```
